```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


```

```python
df = pd.read_csv('/content/anime.csv', encoding='latin-1')
```

```python
#missing values
df.dropna(subset=['genre', 'rating'], inplace=True)
```

```python
df['combined_features'] = df['genre'] + ' ' + df['type'] + ' ' + df['episodes'].astype(str) + ' ' + df['rating'].astype(str) + ' ' + df['members'].astype(str)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
feature_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
# Function to recommend anime
def recommend_anime(target_anime_name, df, cosine_sim, threshold=0.5, top_n=10):
    """
    Recommends anime based on cosine similarity.

    Args:
        target_anime_name (str): Name of the target anime.
        df (pd.DataFrame): DataFrame containing anime data.
        cosine_sim (np.ndarray): Cosine similarity matrix.
        threshold (float): Similarity score threshold for recommendations.
        top_n (int): Number of top recommendations to return.

    Returns:
        list: List of recommended anime names.
    """
    if target_anime_name not in df['name'].values:
        print(f"Anime '{target_anime_name}' not found in the dataset.")
        return []

    target_anime_index = df[df['name'] == target_anime_name].index[0]
    similarity_scores = list(enumerate(cosine_sim[target_anime_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    recommendations = []
    for i, score in similarity_scores[1:]:  # Exclude the target anime itself
        if score >= threshold:
            recommendations.append((df.iloc[i]['name'], score))

            # Sort by score and take top_n
    recommendations.sort(key=lambda x: x[1], reverse=True)
    recommendations = recommendations[:top_n]

    return recommendations
```

```python
# Calculate the cosine similarity matrix using the feature_matrix
cosine_sim = cosine_similarity(feature_matrix, feature_matrix)  # Use feature_matrix instead of combined_features_df

# Example Usage
target_anime = "Naruto"
threshold_value = 0.4  # Experiment with different threshold values
top_n_value = 5 # top 5 anime

recommendations = recommend_anime(target_anime, df, cosine_sim, threshold=threshold_value, top_n=top_n_value) # Now cosine_sim is defined

if recommendations:
  print(f"\nRecommendations for '{target_anime}' (threshold={threshold_value}, top_n={top_n_value}):")
  for anime, score in recommendations:
    print(f"- {anime} (similarity score: {score:.4f})")
else:
  print("No recommendations found for the given anime and threshold.")
```

interview questions

1.Both user-based and item-based collaborative filtering are methods used in recommendation systems to predict a user’s preferences based on their past behavior or the behavior of others. However, they differ in how they utilize this information.

User-Based: Focuses on finding similar users and recommending items based on what those users liked.


Item-Based: Focuses on finding similar items and recommending them based on the user's past preferences."


2.Collaborative filtering is a technique used in recommendation systems where predictions for users are made based on the behavior of other users. It can be user-based or item-based. The system calculates similarity between users or items to recommend new content based on past interactions. For example, it may suggest a movie to a user by finding other users who liked similar movies or by recommending a movie that is similar to one the user has already rated highly.
