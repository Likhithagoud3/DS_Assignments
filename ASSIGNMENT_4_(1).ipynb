```python
import scipy.stats as stats
import numpy as np
```

```python
observed_frequency = np.array([[50, 70],[80, 100],[60, 90],[30, 50],[20, 50]])
```

```python
row_total = np.sum(observed_frequency, axis=1)
col_total = np.sum(observed_frequency, axis=0)
grand_total = np.sum(row_total)
```

```python
expected = np.outer(row_total, col_total) / grand_total
```

```python
chi_squared_statistic = np.sum((observed_frequency - expected)**2 / expected). sum()
df = (observed_frequency.shape[0] - 1) * (observed_frequency.shape[1] - 1)
```

```python
#Critical value for alpha = 0.05
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)
stats.chi2.ppf(1 - alpha, df)
```

```python
if chi_squared_statistic >= critical_value:
    print("Reject H0,There is a relationship between the variables")
else:
    print("fail to reject,the H0, There is no relationship between the variables")
```

```python
print("Observed Values (O): \n", observed_frequency)
print("Expected Values (E): \n", expected)
print("Chi-squared Statistic: ", chi_squared_statistic)
print("Degrees of Freedom: ", df)
print("Critical Value: ", critical_value)
```
