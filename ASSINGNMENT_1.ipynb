```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/sales_data_with_discounts.csv')
```

```python
numerical_columns = dataset.select_dtypes(include=['number'])
```

```python
descriptive_stats = numerical_columns.describe().loc[['mean','50%','std']]
modevalues=numerical_columns.mode().iloc[0]
```

```python
print("Descriptive Statistics:")
print(descriptive_stats)
print("\nMode Values:")
print(modevalues)
```

```python
for col in numerical_columns.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(dataset[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
