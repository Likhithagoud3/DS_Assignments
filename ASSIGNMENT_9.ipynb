```python
# Ensure you exclude non-numeric columns like categorical data if necessary
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns

# Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler() # Use StandardScaler instead of Standardized_data
standardized_data = scaler.fit_transform(df[numerical_features])

# Check standardized data
print("\nStandardized Data (First 5 rows):")
print(standardized_data[:5])
```

```python
# Initialize PCA to retain a certain number of components (e.g., 95% of variance)
from sklearn.decomposition import PCA # Import PCA from sklearn.decomposition
pca = PCA(n_components=0.95)  # Keeps 95% of the explained variance
pca_data = pca.fit_transform(standardized_data)

# View the results
print("\nPCA Transformed Data (First 5 rows):")
print(pca_data[:5])
```

```python
# prompt: Determine the optimal number of principle components using techiniques like screeplot or cumukative explained variance

# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("\nExplained Variance Ratio:")
print(explained_variance_ratio)

# Cumulative Explained Variance
cumulative_variance = np.cumsum(explained_variance_ratio)
print("\nCumulative Explained Variance:")
print(cumulative_variance)

# Scree Plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.show()

# Determine the optimal number of components
# Method 1: Elbow point in the scree plot (look for a point where the variance explained drops off significantly)

# Method 2: Cumulative Explained Variance Threshold (e.g., 95% or 99%)
optimal_components = np.argmax(cumulative_variance >= 0.95) + 1  # Example: 95% variance threshold
print(f"\nOptimal number of components (based on 95% variance): {optimal_components}")

# Method 3: Use a specific number
# You can also set the number of components based on domain knowledge or other criteria
# pca = PCA(n_components=3)
```

```python
# Initialize PCA with the optimal number of components
pca_optimal = PCA(n_components=optimal_components)

# Apply PCA to the standardized data
pca_data_optimal = pca_optimal.fit_transform(standardized_data)

# Print the first few rows of the transformed data
print("\nPCA Transformed Data (Optimal Components - First 5 rows):")
print(pca_data_optimal[:5])
```

```python
# Import KMeans from sklearn.cluster
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(pca_data_optimal)
```

```python
data = df.copy()
data['Cluster'] = kmeans.labels_
```

```python
#Scatter plot of the cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_data_optimal[:, 0], y=pca_data_optimal[:, 1], hue=data['Cluster'], palette='viridis')
```

```python
#Add cluster centers
# Get the number of components from pca_optimal
n_components = pca_optimal.n_components_

# Create a new scaler and fit it on the PCA transformed data
scaler_pca = StandardScaler()
scaler_pca.fit(pca_data_optimal)

# Inverse transform the cluster centers using the new scaler
centroid = scaler_pca.inverse_transform(kmeans.cluster_centers_)

# Now you can plot the centroids
plt.scatter(centroid[:, 0], centroid[:, 1], marker='X', s=200, linewidths=3, color='black')
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Calculate Silhouette Score
silhouette_avg = silhouette_score(pca_data_optimal, kmeans.labels_)
print(f"Silhouette Score: {silhouette_avg}")

# Calculate Davies-Bouldin Index
db_index = davies_bouldin_score(pca_data_optimal, kmeans.labels_)
print(f"Davies-Bouldin Index: {db_index}")
```

```python
# Apply KMeans clustering to the original standardized data
kmeans_original = KMeans(n_clusters=3, random_state=42)
kmeans_original.fit(standardized_data)

# Add cluster labels to the DataFrame
data['Cluster_Original'] = kmeans_original.labels_

# Scatter plot of the clusters for the original data
plt.figure(figsize=(10, 6))
sns.scatterplot(x=standardized_data[:, 0], y=standardized_data[:, 1], hue=data['Cluster_Original'], palette='viridis')
plt.title('KMeans Clustering on Original Standardized Data')
plt.show()

# Calculate Silhouette Score for original data
silhouette_avg_original = silhouette_score(standardized_data, kmeans_original.labels_)
print(f"Silhouette Score (Original Data): {silhouette_avg_original}")

# Apply KMeans clustering to the original standardized data
kmeans_original = KMeans(n_clusters=3, random_state=42)
kmeans_original.fit(standardized_data)

# Add cluster labels to the DataFrame
data['Cluster_Original'] = kmeans_original.labels_

# Scatter plot of the clusters for the original data
plt.figure(figsize=(10, 6))
sns.scatterplot(x=standardized_data[:, 0], y=standardized_data[:, 1], hue=data['Cluster_Original'], palette='viridis')
plt.title('KMeans Clustering on Original Standardized Data')
plt.show()

# Calculate Silhouette Score for original data
silhouette_avg_original = silhouette_score(standardized_data, kmeans_original.labels_)
print(f"Silhouette Score (Original Data): {silhouette_avg_original}")

# Calculate Davies-Bouldin Index for original data
db_index_original = davies_bouldin_score(standardized_data, kmeans_original.labels_)
print(f"Davies-Bouldin Index (Original Data): {db_index_original}")
```
